```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional for better visuals
sns.set(style="whitegrid", palette="pastel")

# Load the merged dataset
df = pd.read_csv('../data/merged_btc_blockchain.csv', parse_dates=['Date'])

# Quick look
df.head()
```

#### Plot Bitcoin Price vs Time

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Bitcoin Close Price', linewidth=2)
plt.title('Bitcoin Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

#### Plot Hash Rate and Difficulty Trends

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

ax[0].plot(df['Date'], df['hash-rate'], label='Hash Rate (EH/s)', color='purple')
ax[0].set_ylabel('Hash Rate')
ax[0].legend()

ax[1].plot(df['Date'], df['difficulty'], label='Mining Difficulty', color='orange')
ax[1].set_ylabel('Difficulty')
ax[1].set_xlabel('Date')
ax[1].legend()

plt.suptitle('Hash Rate & Difficulty Over Time')
plt.tight_layout()
plt.show()
```

#### Correlation Heatmap

```python
# Select numeric columns only
corr = df.drop(columns=['Date']).corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```
