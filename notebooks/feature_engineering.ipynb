```python
import pandas as pd

# Load merged dataset
df = pd.read_csv('../data/merged_btc_blockchain.csv', parse_dates=['Date'])

# Sort just in case
df = df.sort_values('Date')

df.head()
```

```python
df['volatility'] = df['High'] - df['Low']

# 7-day and 30-day moving averages of closing price
df['ma_7'] = df['Close'].rolling(window=7).mean()
df['ma_30'] = df['Close'].rolling(window=30).mean()

# Lag features (shifted closing prices)
df['close_lag_1'] = df['Close'].shift(1)
df['close_lag_3'] = df['Close'].shift(3)
df['hashrate_lag_1'] = df['hash-rate'].shift(1)

df = df.dropna()
df.to_csv('../data/features_btc.csv', index=False)
```
